# Problem 1: Smart Light Controller

```python
# Class to control the smart light
class SmartLight:
    def __init__(self, name):
        # Initialize the light with its name
        self.name = name

    def display_status(self, action):
        # Print the current status based on user action (ON/OFF)
        # using .upper() to ensure the output matches the expected format perfectly
        print(f"\n{self.name} is {action.upper()}")

# Taking inputs from the user
light_name = input("Light Name: ")
action_input = input("Action: ")

# Object creation and method calling
bedroom_light = SmartLight(light_name)
bedroom_light.display_status(action_input)
```

# Problem 2: Employee ID Card System

```python
# Class to store and display employee details
class Employee:
    def __init__(self, name, emp_id, department):
        # Assigning attributes
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        # Displaying the ID card in the requested format
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Taking inputs from the user
emp_name = input("Employee Name: ")
emp_id = input("Employee ID: ")
emp_dept = input("Department: ")

# Object creation and method calling
employee = Employee(emp_name, emp_id, emp_dept)
employee.display_id_card()
```

# Problem 3: Mobile Contact Record

```python
# Class to manage mobile contacts
class ContactRecord:
    def __init__(self, name, phone_number):
        # Storing contact details
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        # Printing the saved contact information
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# Taking inputs from the user
contact_name = input("Contact Name: ")
contact_phone = input("Phone Number: ")

# Object creation and method calling
contact = ContactRecord(contact_name, contact_phone)
contact.display_contact()
```

# Problem 4: Product Price Tag Generator

```python
# Class for generating product price tags
class Product:
    def __init__(self, name, price):
        # Initialize product attributes
        self.name = name
        self.price = price

    def print_price_tag(self):
        # Display formatted price tag with the ₹ symbol
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")

# Taking inputs from the user
product_name = input("Product Name: ")
product_price = input("Price: ")

# Object creation and method calling
item = Product(product_name, product_price)
item.print_price_tag()
```

# Problem 5: Movie Rating Display System

```python
# Class to handle movie ratings
class Movie:
    def __init__(self, name, rating):
        # Store movie name and rating
        self.name = name
        self.rating = rating

    def display_details(self):
        # Output the formatted movie rating
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# Taking inputs from the user
movie_name = input("Movie Name: ")
movie_rating = input("Rating: ")

# Object creation and method calling
movie = Movie(movie_name, movie_rating)
movie.display_details()
```

# Problem 6: Delivery Address Manager

```python
# Class to manage and display customer delivery details
class DeliveryManager:
    def __init__(self, customer_name, address):
        # Storing delivery info
        self.customer_name = customer_name
        self.address = address

    def print_details(self):
        # Printing formatted delivery details
        print("\nDelivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Taking inputs from the user
customer_name = input("Customer Name: ")
customer_address = input("Address: ")

# Object creation and method calling
delivery = DeliveryManager(customer_name, customer_address)
delivery.print_details()
```
